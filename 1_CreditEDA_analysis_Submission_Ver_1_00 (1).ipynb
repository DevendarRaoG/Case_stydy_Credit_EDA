# OCCUPATION TYPE analysis 

```python
# Definition used for showing values on the plots
def show_values_on_bars(axs):
    def _show_on_single_plot(ax):        
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height()
            value = '{:.0f}'.format(p.get_height())
            ax.text(_x, _y, value, ha="center") 

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)

# Plotting which category of profession applied for maximum loans
# for Target 0 and Target 1.

from collections import Counter



# TARGET - 0 CALCULATIONS
professionCounts = Counter(list(dfApplicationData_TargetZero.OCCUPATION_TYPE))

# Excluding "NOT_PROVIDED" column as this is user created
professionCounts.pop('NOT_PROVIDED', None)

dfProfessionCountsZero = pd.DataFrame(list(professionCounts.values()), index = professionCounts.keys())


# TARGET - 1 CALCULATIONS
professionCounts = Counter(list(dfApplicationData_TargetOne.OCCUPATION_TYPE))

# Excluding "NOT_PROVIDED" column as this is user created
professionCounts.pop('NOT_PROVIDED', None)

dfProfessionCountsOne = pd.DataFrame(list(professionCounts.values()), index = professionCounts.keys())
```

```python

plt.figure(figsize = (14,8))
plt.figure(1)
sns.set_style("whitegrid")


ax = plt.subplot(121)
sns.barplot(x = dfProfessionCountsZero.sort_values(by = 0, ascending = False).index, y = dfProfessionCountsZero.sort_values(by = 0, ascending = False)[0])
plt.title('Target - 0 - OCCUPATION TYPE Analysis')
plt.ylabel('Number of applications')
plt.xlabel('OCCUPATION TYPE')
show_values_on_bars(ax)
plt.xticks(rotation = 90)

ax = plt.subplot(122)
sns.barplot(x = dfProfessionCountsOne.sort_values(by = 0, ascending = False).index, y = dfProfessionCountsOne.sort_values(by = 0, ascending = False)[0])
plt.title('Target - 1 - OCCUPATION TYPE Analysis')
plt.ylabel('Number of applications')
plt.xlabel('OCCUPATION TYPE')
show_values_on_bars(ax)
plt.xticks(rotation = 90)
```

# Age based analysis for Target 0 and Target 1

```python
# Creating brackets of number of days to analyse the Target 0 and Target 1

# For TARGET 0
# Making a proper copy of dfApplicationData_TargetZero (dataframe) as this is slice of master df
dfApplicationData_TargetZero = dfApplicationData_TargetZero.copy()

# Considering 365 per year to approx the number of years applicant has just before application submission
dfApplicationData_TargetZero['APPLICANT_AGE_REL_TO_APPLICATION'] = dfApplicationData_TargetZero.DAYS_BIRTH.apply(lambda x: int(abs(x)/365))

# For Target 1
dfApplicationData_TargetOne = dfApplicationData_TargetOne.copy()
dfApplicationData_TargetOne['APPLICANT_AGE_REL_TO_APPLICATION'] = dfApplicationData_TargetOne.DAYS_BIRTH.apply(lambda x: int(abs(x)/365))
```

```python
plt.figure(figsize = (16,8))
plt.figure(1)

ax = plt.subplot(121)
dfBinnedValues = pd.cut(dfApplicationData_TargetZero.APPLICANT_AGE_REL_TO_APPLICATION, bins = range(0,100, 5), include_lowest = True)
ax = dfBinnedValues.value_counts(sort=False).plot.bar(rot=0, color='g')      # Here each intervals total coming from df.value_counts function call
plt.xticks(rotation = 90)
ax.set_xticklabels([str(c).replace(',', ' to') for c in dfBinnedValues.cat.categories])
plt.title('Target 0 - Age Based number of applications')
plt.ylabel('Total Number of Applications')
plt.xlabel('Applicant Age Range')
show_values_on_bars(ax)

ax = plt.subplot(122)
dfBinnedValues = pd.cut(dfApplicationData_TargetOne.APPLICANT_AGE_REL_TO_APPLICATION, bins = range(0,100, 5), include_lowest = True)
ax = dfBinnedValues.value_counts(sort=False).plot.bar(rot=0, color='r')      # Here each intervals total coming from df.value_counts function call
plt.xticks(rotation = 90)
ax.set_xticklabels([str(c).replace(',', ' to') for c in dfBinnedValues.cat.categories])
plt.title('Target 1 - Age Based number of applications')
plt.ylabel('Total Number of Applications')
plt.xlabel('Applicant Age Range')
show_values_on_bars(ax)
```

```python
# Definition used for showing values on the plots
def show_values_on_bars_vertical(axs):
    def _show_on_single_plot(ax):        
        for p in ax.patches:
            _x = p.get_x() + p.get_width() / 2
            _y = p.get_y() + p.get_height()
            value = '{:.0f}'.format(p.get_height())
            ax.text(_x, _y, value, ha="center", rotation = 90) 

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
```
